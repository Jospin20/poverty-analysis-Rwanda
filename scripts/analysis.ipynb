```python
# List of numeric columns we expect
num_cols = ['EICV7_2024', 'EICV5_2017', 'Change_points', 'Change_percent', 'Lower_CI_2024', 'Upper_CI_2024']

# Clean and convert to float
for col in num_cols:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
```

```python
# Sort by Upper_CI_2024 to spot the outlier
df_clean[['Province', 'Upper_CI_2024']].sort_values(by='Upper_CI_2024', ascending=False)
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode Province names
le = LabelEncoder()
df_clean['Province_encoded'] = le.fit_transform(df_clean['Province'])

# Show updated table
df_clean[['Province', 'Province_encoded']]
```

```python
from sklearn.preprocessing import StandardScaler

# Select only numeric columns (excluding target for now)
features_to_scale = ['EICV7_2024', 'EICV5_2017', 'Change_points', 'Change_percent', 'Lower_CI_2024', 'Upper_CI_2024']
scaler = StandardScaler()

# Create scaled DataFrame
scaled_features = scaler.fit_transform(df_clean[features_to_scale])
df_scaled = pd.DataFrame(scaled_features, columns=[col + '_scaled' for col in features_to_scale])

# Merge with the original dataset
df_model = pd.concat([df_clean, df_scaled], axis=1)

# Preview final dataset
df_model.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# We'll use the scaled features
X = df_model[[col + '_scaled' for col in ['EICV7_2024', 'EICV5_2017', 'Change_points', 'Change_percent', 'Lower_CI_2024', 'Upper_CI_2024']]]

# Elbow method: test k from 1 to 10
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot
plt.figure(figsize=(8, 4))
plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method to Choose Optimal K')
plt.grid(True)
plt.show()
```

```python
K_range = range(1, min(7, len(X)+1))  # max clusters = number of samples

inertia = []
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method (max clusters limited to number of samples)')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

k = 3  # or your chosen number
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
kmeans.fit(X)

# Add cluster labels to your dataset
df_model_final = df_model_final.loc[X.index]  # align indices just in case
df_model_final['Cluster'] = kmeans.labels_

# Show cluster assignment by Province
result = df_model_final[['Province', 'Cluster']]
print(result)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Perform PCA to reduce dimensionality to 2D for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8,6))
scatter = plt.scatter(X_pca[:,0], X_pca[:,1], c=df_model_final['Cluster'], cmap='Set1', s=100)

# Annotate points with province names
for i, province in enumerate(df_model_final['Province']):
    plt.annotate(province, (X_pca[i,0]+0.1, X_pca[i,1]))

plt.title('Province Clusters Based on Poverty Indicators (PCA Projection)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(scatter, label='Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X, df_model_final['Cluster'])
print(f'Silhouette Score for k={k}: {score:.3f}')
```

```python
output_file = 'poverty_province_clusters.csv'
df_model_final.to_csv(output_file, index=False)
print(f"Data exported to {output_file}")
```

## Clustering Analysis Summary

- We applied K-Means clustering on scaled poverty indicators across Rwandas provinces.
- The elbow method suggested **k=3** clusters as optimal.
- Cluster 0 groups provinces with higher poverty rates (South, West, East, Rwanda).
- Cluster 1 contains the North province, indicating distinct poverty trends.
- Cluster 2 is Kigali City, reflecting urban characteristics.
- Silhouette Score for k=3 is approximately 0.350, indicating moderate separation between clusters. This suggests some overlap but the clusters still provide useful insights into poverty trends across provinces.
- This clustering provides strategic insights for targeted poverty alleviation efforts.
